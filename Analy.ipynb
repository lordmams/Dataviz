```python
import csv

def read_dates_from_csv(csv_file):
    dates = []
    with open(csv_file, 'r') as file:
        reader = csv.reader(file, delimiter=';')
        next(reader)
        for row in reader:
            start_date = row[0]
            end_date = row[1]
            dates.append((start_date, end_date))
    return dates




csv_file = 'mouvements-sociaux-depuis-2002.csv' 
dates = read_dates_from_csv(csv_file)
print("Dates from CSV file:")
for date in dates:
    print(date)
```

```python
import pandas as pd

def format_date_column(csv_file, date_column):
    df = pd.read_csv(csv_file, delimiter=';')
    
    df[date_column] = pd.to_datetime(df[date_column]).dt.strftime('%Y-%m-%d')
    return df

csv_file = 'consommation-quotidienne-brute.csv'  # Préciser le chemin du dataset
date_column = 'Date'  # Spéfier le nom de la colonne date
formatted_df = format_date_column(csv_file, date_column)
```

```python
def add_movement_social_column(formatted_df, dates, date_column):
    formatted_dates = []
    for start_date, end_date in dates:
        start_date = pd.to_datetime(start_date)
 
        end_date = pd.to_datetime(end_date) if end_date else start_date
        formatted_dates.append((start_date, end_date))
    formatted_df[date_column] = pd.to_datetime(formatted_df[date_column])
    current_date = pd.to_datetime('today').normalize()
    
   
    formatted_df['mouvement_social'] = False
  
    for index, row in formatted_df.iterrows():
    
        row_date = row[date_column]
        
        
        for start_date, end_date in formatted_dates:
            if start_date <= row_date <= end_date:
                formatted_df.at[index, 'mouvement_social'] = True
                break 
    
    return formatted_df


formatted_df_with_movement_social = add_movement_social_column(formatted_df, dates, date_column)

output_csv_file = 'Consomation&Mouvement.csv'
formatted_df_with_movement_social.to_csv(output_csv_file, index=False)
```
